### Create 2  New Columns

```python
EPL_Dataset['Minspermatch']=(EPL_Dataset['Mins']/EPL_Dataset['Matches']).astype(int)
EPL_Dataset['Goalspermatch']=(EPL_Dataset['Goals']/EPL_Dataset['Matches']).astype(float)
EPL_Dataset.head()
```

```python
# Total Goals in 2020-2021 Season
toltal_goal=EPL_Dataset['Goals'].sum()
print(toltal_goal)
```

```python
# Total Penalty Goals in 2020-2021 Season
Penalty_Goals=EPL_Dataset['Penalty_Goals'].sum()
print(Penalty_Goals)
```

```python
# Total Penalty Attempted in 2020-2021 Season
total_Penalty_Attempted=EPL_Dataset['Penalty_Attempted'].sum()
print(total_Penalty_Attempted)
```

```python
# Pie chart for penalties missed vs Scored
plt.figure(figsize=(13,6))
Penalty_not_score=EPL_Dataset['Penalty_Attempted'].sum()-Penalty_Goals
data=[Penalty_not_score,total_Penalty_Attempted]
labels=['Penalties missed','Penalties Score']
color=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
# Most Players from which countries
nationality=EPL_Dataset.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(13,6),color=sns.color_palette("magma"))
plt.show()
```

```python
# Clubs with the maximum players in their squad
EPL_Dataset['Club'].value_counts().nlargest(5).plot(kind='bar',color=sns.color_palette("viridis"))
plt.show()
```

```python
# Clubs with the Least players in their squad
EPL_Dataset['Club'].value_counts().nsmallest(5).plot(kind='bar',color=sns.color_palette("viridis"))
plt.show()
```

### Players based on Age group

```python
under20=EPL_Dataset[EPL_Dataset['Age']<=20]
age20_25=EPL_Dataset[(EPL_Dataset['Age']>20) & (EPL_Dataset['Age']<=25)]
age25_30=EPL_Dataset[(EPL_Dataset['Age']>25) & (EPL_Dataset['Age']<=30)]
above30=EPL_Dataset[EPL_Dataset['Age']>30]
```

```python
x=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels=["<=20",">20 & <=25 ",">25 & <=25",">30"]
plt.title("Total Players with Age", fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 players in each other 
players_under_20=EPL_Dataset[EPL_Dataset['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("cubehelix"))
plt.show()
```

```python
# Under 20 Players in manchester united
players_under_20[players_under_20["Club"]=='Manchester United']
```

```python
# Under 20 Players in Chelsea
players_under_20[players_under_20["Club"]=='Chelsea']
```

### Average age of players in each club

```python
plt.figure(figsize=(13,6))
sns.boxplot(x='Club',y='Age',data=EPL_Dataset)
plt.xticks(rotation=90)
plt.show()
```

```python
num_player=EPL_Dataset.groupby('Club').size()
data=(EPL_Dataset.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)   
```

```python
 # TOTAL ASSISTS FROM EACH OTHER
assists_by_clubs=pd.DataFrame(EPL_Dataset.groupby('Club',as_index=False)['Assists'].sum())  
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=assists_by_clubs.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=70)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists ',fontsize=20)
plt.show()
```

```python
# Top 10 Assists
top_10_assists=EPL_Dataset[["Name","Club","Assists","Matches"]].nlargest
top_10_assists
```

```python
Goals_by_clubs=pd.DataFrame(EPL_Dataset.groupby('Club',as_index=False)['Goals'].sum())  
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=70)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals ',fontsize=20)
plt.show()
```

```python
# Top 10 Goals
top_10_goals=EPL_Dataset[["Name","Club","Goals","Matches"]].nlargest(n=10,columns='Goals')
top_10_assists
```

```python
# Goals per match
top_10_goals_per_match=EPL_Dataset[["Name","Goalspermatch","Goals","Matches"]].nlargest(n=10,columns='Goalspermatch')
top_10_assists
```

```python
# Pie chart for penalties missed vs Scored
plt.figure(figsize=(13,6))
assists=EPL_Dataset['Assists'].sum()
data=[toltal_goal-assists,assists]
labels=['Goals without Assists','Goals with Assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards
epl_yellow=EPL_Dataset.sort_values(by="Yellow_Cards",ascending=False)[:10]
plt.figure(figsize=(13,6))
plt.title('Players with Most Yellow Cards ',fontsize=20)
c=sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color="yellow")
plt.ylabel('Number of yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
plt.show()
```
